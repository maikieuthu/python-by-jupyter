```python
import requests
import bs4
re = requests.get("https://quotes.toscrape.com/")
soup=bs4.BeautifulSoup(re.text,"lxml")
soup
```

```python
#Get the name of authors on the first page by using Set
soup.select(".author")
```

```python
authors= set()
for n in soup.select(".author"):
    authors.add(n.text)
authors
   
```

```python
#Get the quotes on the first page by using List
soup.select(".text")
quotes=list()
for i in soup.select(".text"):
    quotes.append(i.text)
quotes
```

```python
#Get the tag on the right side of the website
soup.select(".tag-item")
for items in soup.select(".tag-item"):
    print(items.text)
```

```python
#Get the name of authors form 1st to 10th page by using List:
url = "https://quotes.toscrape.com/page/"
authors= list()

for i in range (1,10):
    scrape_url=url+str(i)
    res= requests.get(scrape_url)
    soup=bs4.BeautifulSoup(res.text,"lxml")
    for n in soup.select(".author"):
       authors.append(n.text)
    
authors
```
